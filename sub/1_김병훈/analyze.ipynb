### Req 2-4. 리뷰 개수 기준 이용자 정렬

```python
# users 데이터 프레임과 reviews 데이터 프레임 
data["users"].head(5)
```

```python
data["reviews"].head(5)
```

```python
users_reviews = pd.merge(data["users"], data["reviews"], left_on="id", right_on="user")
users_reviews
```

```python
users_reviews_grouped = users_reviews.groupby(
    ["user", "gender", "age"]
).user.count().reset_index(name="review_count")
```

```python
users_reviews_sorted = users_reviews_grouped.sort_values(["review_count"], ascending=False).reset_index()
users_reviews_sorted
```

```python
import json
import pandas as pd
```

```python
with open('./movies_kr.json', 'r') as f:
    data = json.loads(f.read())
```

```python
df = pd.json_normalize(data)
df = df.fillna(0)
```

```python
df.head()
```

```python
for idx, row in df.iterrows():
    if row['fields.provider.buy'] != 0:
        print(row['pk'])
        print(row['fields.provider.buy'][0]['provider_name'])    
    if row['fields.provider.rent'] != 0:
        print(row['fields.provider.rent'][0]['provider_name'])
    if row['fields.provider.flatrate'] != 0:
        print(row['fields.provider.flatrate'][0]['provider_name'])
```

```python
for idx, row in df.iterrows():
    if row['fields.provider.rent'] != 0:
        print(f"{idx} : {len(row['fields.provider.rent'])}")
```

```python
def provider_list_to_set(p_set, p_list):
    size = len(p_list)
    for idx in range(0, size):
        # print(p_list[idx])
        p_set.add(p_list[idx]['provider_name'])
```

```python
for idx, row in df.iterrows():
    provider_set = set()
    if row['fields.provider.buy'] != 0:
        provider_list_to_set(provider_set, row['fields.provider.buy'])
    if row['fields.provider.rent'] != 0:
        provider_list_to_set(provider_set, row['fields.provider.rent'])
    if row['fields.provider.flatrate'] != 0:
        provider_list_to_set(provider_set, row['fields.provider.flatrate'])    
    print(provider_set)
```
